#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt

#Carregando os dados do arquivo
dada_sus = pd.read_csv('b1abb33c4f.csv')

#Verificando se os dados foram carregados corretamente
dada_sus.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
#Analisando aos valores das variaveis
dada_sus[['APGAR1','APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']]
```

```python
plt.rc('figure', figsize=(10, 10))
fig, axes = plt.subplots(3, 2)

sns.countplot(ax = axes[0, 0], x='APGAR1', data=dada_sus)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=dada_sus)
sns.countplot(ax = axes[1, 0], x='SEXO', data=dada_sus)
sns.countplot(ax = axes[1, 1], x='ESTCIVMAE', data=dada_sus)
sns.countplot(ax = axes[2, 0], x='GESTACAO', data=dada_sus)

axes[2, 0].tick_params(axis="x", rotation = 30)
plt.subplots_adjust(wspace=0.3, hspace=0.7)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')

plt.rc('figure', figsize=(10, 8))

fig, ax1 = plt.subplots()

sns.barplot(ax = ax1, x='APGAR5', y='CONSULTAS', data=dada_sus)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='SEXO', data=dada_sus)
sns.countplot(ax = axes[0, 1], x='GESTACAO', data=dada_sus)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=dada_sus)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=dada_sus)

axes[0, 1].tick_params(axis="x", rotation = 30)
plt.subplots_adjust(wspace=0.3, hspace=0.7)
```
